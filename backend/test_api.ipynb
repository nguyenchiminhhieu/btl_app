```python
import requests
import json
import base64

def transcribe_audio(api_key, audio_path):
    url = "https://speech.googleapis.com/v1p1beta1/speech:recognize?key=" + api_key

    with open(audio_path, 'rb') as audio_file:
        audio_content = audio_file.read()

    # M√£ h√≥a d·ªØ li·ªáu √¢m thanh th√†nh base64
    audio_content_base64 = base64.b64encode(audio_content).decode("utf-8")

    # Chu·∫©n b·ªã payload cho y√™u c·∫ßu
    data = {
        "config": {
            "encoding": "LINEAR16",
            "sampleRateHertz": 44100,
            "languageCode": "en-US"
        },
        "audio": {
            "content": audio_content_base64
        }
    }

    # G·ª≠i y√™u c·∫ßu POST t·ªõi API Google Speech-to-Text
    response = requests.post(url, json=data)

    if response.status_code == 200:
        result = response.json()
        for r in result["results"]:
            print("Transcript:", r["alternatives"][0]["transcript"])
    else:
        print("Error:", response.status_code, response.text)
# Test
api_key = "AIzaSyDBiXRGObWD5gv4cCHhIbuvvM4a6ewr1Hk"
transcribe_audio(api_key, "./voice_preview_aria.wav")
```

```python
import requests
import json

import requests
import json
import base64

def synthesize_text(api_key, text, output_audio_path, sample_rate=24000):
    url = f"https://texttospeech.googleapis.com/v1/text:synthesize?key={api_key}"

    headers = {
        "Content-Type": "application/json",
    }

    data = {
        "input": {"text": text},
        "voice": {
            "languageCode": "en-US",
            "ssmlGender": "FEMALE"
        },
        "audioConfig": {
            "audioEncoding": "MP3",
            "sampleRateHertz": sample_rate  # C√≥ th·ªÉ ƒëi·ªÅu ch·ªânh sampleRate ·ªü ƒë√¢y
        }
    }

    response = requests.post(url, headers=headers, json=data)

    if response.status_code == 200:
        audio_content = response.json()["audioContent"]
        
        # Gi·∫£i m√£ base64 v√† ghi v√†o t·ªáp
        audio_data = base64.b64decode(audio_content)
        
        with open(output_audio_path, "wb") as out:
            out.write(audio_data)
    else:
        print("Error:", response.status_code, response.text)

api_key = "AIzaSyCC6sL6xqDI3wNUC15c3FRfOip_6xqQ7PY"
# Test
synthesize_text(api_key, "Hello, this is a test.", "output_audio.mp3")
```
